```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
from IPython.display import Image, display
from dotenv import load_dotenv
import os
from langchain_google_genai import ChatGoogleGenerativeAI
load_dotenv()
```

```python
model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0,
    max_tokens=None,
    timeout=None,
    max_retries=2,
    # other params...
)
```

```python
class Blogstate(TypedDict):
    title: str
    outline: str
    blog_content: str
```

```python
graph = StateGraph(Blogstate)
```

```python
def create_outline(state:Blogstate)->Blogstate:
    title = state["title"]
    prompt = f"Create a detailed outline based on the title: {title}"
    response = model.invoke(prompt)
    state["outline"] = response.content
    return state
```

```python
def create_blog(state:Blogstate)->Blogstate:
    outline = state["outline"]
    title = state["title"]
    prompt = f"Write a comprehensive blog on title - {title} use the following outline: {outline}"
    response = model.invoke(prompt)
    state["blog_content"] = response.content
    return state
```

```python
graph.add_node("create_outline", create_outline)
graph.add_node("create_blog", create_blog)
graph.add_edge(START, "create_outline")
graph.add_edge("create_outline", "create_blog")
graph.add_edge("create_blog", END)
workflow = graph.compile()
```

```python
initilize_state = {"title": "The Future of AI in Healthcare"}
final_state = workflow.invoke(initilize_state)
print(final_state["blog_content"])
```
